4.Data Preprocessing
Short Explanation: Data preprocessing is a crucial step in machine learning that prepares raw data for model training. In this phase, the dataset is cleaned and transformed to ensure that machine learning algorithms can process it correctly. This includes handling missing values, converting categorical variables into numerical form, and scaling numerical features so that all variables are on a similar scale.


4.1 : Handling Missing Values
Missing values can negatively affect machine learning models. Therefore, it is important to detect and handle them before training. Numerical missing values are replaced using the median, while categorical missing values are replaced using the most frequent category.


4.1.1: Check for Missing Values

```python
df.isnull().sum()
```

4.1.2: Fill Missing Values:

```python
# Separate numeric and categorical columns
num_cols = df.select_dtypes(include="number").columns.tolist()
cat_cols = df.select_dtypes(include="object").columns.tolist()

# Fill numeric columns with median
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# Fill categorical columns with mode
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Verify no missing values remain
df.isnull().sum().sum()
```

4.2 Encoding Categorical Variables
Explanation:

Machine learning algorithms require numerical input. Therefore, categorical (text) variables must be converted into numerical format. One-Hot Encoding is used to transform each category into a binary feature.

5.Machine learning Models:

In this section, five regression algorithms are trained to predict the target variable (total_sales_price). The models are compared using evaluation metrics such as MAE, RMSE, and R² to identify the best-performing approach.

Import+Evaluation Function:

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def evaluate_regression(y_true, y_pred, model_name="Model"):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    return {"Model": model_name, "MAE": mae, "RMSE": rmse, "R2": r2}
```

Create a place to store results:

```python
results = []
```

5.1-Linear Regression:
Linear Regression is used as a baseline regression model. It assumes a linear relationship between the input features and the target variable.

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)
results.append(evaluate_regression(y_test, y_pred_lr, "Linear Regression"))
```

```python
lr_result = evaluate_regression(y_test, y_pred_lr, "Linear Regression")
print(lr_result)
```

Lets do a small comparison:

```python
comparison = pd.DataFrame({
    "Actual": y_test.values[:10],
    "Predicted": y_pred_lr[:10]
})

comparison
```

conclusion:Linear Regression was used as a baseline model to predict the total sales price. The comparison between actual and predicted values shows that the model is able to capture the general trend in the data; however, it produces large prediction errors for several observations. In some cases, the model underestimates or overestimates the actual values significantly, and it even generates negative predictions, which are not meaningful for sales prices.
These results indicate that the relationship between the features and the target variable is not purely linear. 5.2-Decision Tree Regressor:

Decision Tree Regressor models non-linear relationships by splitting the data into decision rules. It can capture complex patterns but may overfit.

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(random_state=42)
dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)

results.append(evaluate_regression(y_test, y_pred_dt, "Decision Tree"))
print(evaluate_regression(y_test, y_pred_dt, "Decision Tree"))
```

lets show com:

```python
import pandas as pd

dt_compare = pd.DataFrame({
    "Actual": y_test.values[:10],
    "Predicted": y_pred_dt[:10]
})
dt_compare
```

The Decision Tree model shows near-perfect predictions, indicating strong learning capability. However, the perfect match between actual and predicted values suggests overfitting, which may limit its ability to generalize to new data.
This behavior is normal for Decision Trees when:
•	No depth limit is set
•	Data is complex
•	Many features exist
That’s why Random Forest and Gradient Boosting are used next.


5.3-Random Forest Regressor:

Random Forest is an ensemble method that combines multiple decision trees to improve accuracy and reduce overfitting.

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=200, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)

results.append(evaluate_regression(y_test, y_pred_rf, "Random Forest"))
print(evaluate_regression(y_test, y_pred_rf, "Random Forest"))
```

Code to show the run (Actual vs Predicted)

```python
rf_compare = pd.DataFrame({
    "Actual": y_test.values[:10],
    "Predicted": y_pred_rf[:10]
})
rf_compare
```

Conclusion:Random Forest achieved highly accurate predictions with very small errors. By combining multiple trees, it reduces overfitting compared to a single Decision Tree and provides strong generalization for predicting total sales price.

5.4-Gradient Boosting Regressor:

Gradient Boosting builds models sequentially, where each model corrects the errors of the previous one. It often achieves strong performance on structured data.

```python
from sklearn.ensemble import GradientBoostingRegressor

gb = GradientBoostingRegressor(random_state=42)
gb.fit(X_train, y_train)

y_pred_gb = gb.predict(X_test)

results.append(evaluate_regression(y_test, y_pred_gb, "Gradient Boosting"))
print(evaluate_regression(y_test, y_pred_gb, "Gradient Boosting"))
```

Code to show the run (Actual vs Predicted)

```python
gb_compare = pd.DataFrame({
    "Actual": y_test.values[:10],
    "Predicted": y_pred_gb[:10]
})
gb_compare
```

conclusion:Gradient Boosting achieves accurate and stable predictions by iteratively correcting errors from previous models. The close alignment between actual and predicted values indicates strong learning capability and good generalization performance.

7. Results and Discussion
7.1 Results

The performance of the five regression models was evaluated using Mean Absolute Error (MAE), Root Mean Squared Error (RMSE), and the coefficient of determination (R²). The results clearly demonstrate notable differences in predictive accuracy among the models.

The Random Forest Regressor achieved the best overall performance, recording the lowest MAE, lowest RMSE, and the highest R² score (≈ 0.999). This indicates that Random Forest is highly effective in predicting the total sales price with minimal error. The Gradient Boosting Regressor also showed excellent performance, with results very close to those of Random Forest, confirming its strong predictive capability.

The Decision Tree Regressor produced very accurate predictions on the observed data; however, its performance suggested potential overfitting, as it closely matched the actual values without sufficient generalization. Linear Regression showed significantly higher error values and a much lower R² score, indicating that it failed to model the complex relationships within the data. Support Vector Regression (SVR) required extensive computational time and achieved lower accuracy compared to ensemble-based models.

Overall, the results confirm that ensemble learning methods outperform single and linear models for this prediction task.

7.2 Discussion

The superior performance of Random Forest and Gradient Boosting can be attributed to their ability to model complex, non-linear relationships and reduce overfitting through ensemble techniques. By combining multiple weak learners, these models improve robustness and generalization, which is particularly important given the skewed nature of the total sales price and the presence of outliers in the dataset.

In contrast, Linear Regression assumes a linear relationship between features and the target variable, which does not hold true for this dataset, resulting in poor predictive performance. Although the Decision Tree model demonstrated high accuracy, its tendency to overfit limits its reliability when applied to unseen data. SVR, while capable of capturing non-linear patterns, proved inefficient for this dataset due to high computational cost and reduced accuracy after scaling and encoding.

These findings highlight the importance of selecting appropriate models based on data characteristics. For large, high-dimensional datasets with complex relationships, ensemble-based regression models provide a more reliable and effective solution.

Final Insight:

Based on both quantitative evaluation metrics and qualitative analysis, Random Forest Regressor is identified as the most suitable model for predicting total sales price in this project, followed closely by Gradient Boosting Regressor.
